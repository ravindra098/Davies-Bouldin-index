```
#REMOVING MISSING VALUES
df =df.dropna()
```

```
#MAKING A COPY OF THE OBJECT'S INDICES AND DATA.
X1= df.copy()
```

```
del X1['Customer']
```

```
del X1['Effective To Date']
```

```
X4=pd.get_dummies(X1)
```

```
n=10
```

```
clf=k_means(X4,n_clusters=n)
```

```
clf
```

```
centroids=clf[0]
```

```
labels=clf[1]
```

```
centroids
```

```
labels
```

```
#Computing the Davies-Bouldin score.
#The score is defined as the average similarity measure of each cluster with its most similar cluster, where similarity is the ratio of within-cluster distances to between-cluster distances. Thus, clusters which are farther apart and less dispersed will result in a better score.
#Here the minimum score is zero, with lower values indicating better clustering.
import sklearn as sklearn
sklearn.metrics.davies_bouldin_score(X4, labels)
```
